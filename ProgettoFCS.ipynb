```python
%%time
import numpy as np
import pandas as pd
import re
import math
import time
doc=pd.read_csv('googleplaystore.csv')
doc=doc.drop(doc.index[10472]).reset_index(drop=True)    #alla riga 10472 ho una row da eliminare
display(doc)
```

```python
start = time.time()
```

# 1) Convert app size into number

```python
%%time

#definisco il pattern da cercare
m = re.compile('^(?P<number>\d+[.]*\d*)(?P<unit>[M|k])$')

#definisco la funzione per la conversione
def size2num(x):
    m1 = m.match(x)
    if m1 is None:
        return np.nan
    num = float(m1.group('number'))    
    uni = str(m1.group('unit'))
    if uni == 'M':
        return int(np.round(num * 10**6))
    if uni == 'k':
        return int(np.round(num * 10**3))
    
#applico la funzione size2num alla colonna Size
doc['Size'] = doc['Size'].apply(size2num)
display(doc)
```

# 2) Convert the number of installs to a number

```python
%%time

m = re.compile('^(?P<number>\d+([,]\d*)*)[+]$')

def inst2num(x):
    m1 = m.match(x)
    if m1 is None:
        return np.nan
    num = str(m1.group('number'))   
    return int(num.replace(',', ''))

doc['Installs'] = doc['Installs'].apply(inst2num)
display(doc)
```

# 3) Transform “Varies with device” into a missing value

```python
%%time

doc = doc.replace('Varies with device', np.nan)
display(doc)
```

# 4) Convert Current Ver and Android Ver into a dotted number (e.g. 4.0.3 or 4.2)

```python
%%time

#definisco la lista a contenente tutti gli elementi della colonna 'Current Ver' come stringhe
a = [str(i) for i in doc['Current Ver']]
#sfrutto la list comprehension: re.findall restituisce il pattern indicato nel suo argomento; se lo trova con re.search.
#se non lo trova restituisco nan.
lst = [re.findall(r'\d+[.][.|\d]*',x)[-1].strip('.') if re.search(r'\d+[.][.|\d]*',x) else np.nan for x in a ]
doc['Current Ver'] = lst

#stessa cosa per 'Android Ver'
a = [str(i) for i in doc['Android Ver']]
lst = [re.findall(r'\d+[.][.|\d]*',x)[-1].strip('.') if re.search(r'\d+[.][.|\d]*',x) else np.nan for x in a ]
doc['Android Ver'] = lst
display(doc)
```

# 5) Remove the duplicates

```python
%%time

doc2=doc.groupby(['App']).size().reset_index(name='counts')  #mi da il numero delle occorrenze per ogni App

display(doc2[doc2['counts'] >1])    #printo tutte le righe che hanno almeno due occorrenze per nome di App
                                                   
# 8 Ball Pool ha ad es. 7 occorrenze, quello che cambia sono le reviews e in un caso anche Category.        
display(doc[doc["App"] == '8 Ball Pool'])
```

Per considerare i duplicati, farei riferimento al nome App. Gli altri attributi variano da verisone a ver. Infatti, ad esempio Current Ver e Android Ver sono spesso nan e quindi non utili; Category spesso varia tra duplicati; ugualmente varia Reviews, cosa che teoricamente dovrebbe influenzare anche Rating...

Sui duplicati devo inoltre porre la condizione di tenere quelli con un n maggiore di Reviews.

```python
#converto le Reviews in  numero
doc.Reviews = doc.Reviews.apply(pd.to_numeric)
#sto ordinando per Reviews -> in caso di duplicati, quello con maggior Reviews sarà in fondo.
doc=doc.sort_values(['Reviews'])             
display(doc.dtypes)

doc=doc.drop_duplicates(subset=['App'], keep = 'last')
#tenere l'ultimo mi assicura che tengo quello con maggiori Reviews
doc = doc.sort_index()
doc = doc.reset_index(drop=True) #resetto l'indice per evitare 'salti'
display(doc)
```

# 8) Create two dataframes: one for the genres and one bridging apps and genres. So that, for instance, the app Pixel Draw - Number Art Coloring Book appears twice in the bridging table, once for Art & Design, once for Creativity

```python
%%time

#creo il df dei Generi.
#Split divide ogni valore di 'Genres' che contiene un ; in due valori salvati in due colonne diverse.
gen = pd.DataFrame([genres.split(';') for genres in doc.Genres])
gen.columns = ['gen1', 'gen2']
display(gen)

###############################################

#ora creo la bridging table di App più i generi
app = pd.DataFrame(doc['App'])

#applico un merge prima e poi un melt
#merge unisce app con gen; melt poi crea due righe nel caso ci siano due generi per una App.
bg = pd.merge(app, gen,left_index=True,right_index=True).melt(id_vars = ['App'], value_name = "Genres")
bg=bg.drop(columns = "variable").dropna().reset_index(drop=True)
display(bg)

display(bg[bg["App"] == "Pixel Draw - Number Art Coloring Book"]) #la richiesta è soddisfatta.
```

# 9) For each genre, create a new column of the original dataframe. The new columns must have boolean values (True if the app has a given genre)

```python
%%time

#creo doc3 concatenando a doc le dummy variables delle colonne gen1 e gen2 del dataframe dei generi.
doc3 = pd.concat([doc, pd.get_dummies(gen["gen1"], dtype = "bool")], axis= 1)
doc3 = pd.concat([doc3, pd.get_dummies(gen["gen2"], dtype = "bool")], axis= 1).drop(columns='Genres')

pd.set_option('display.max_columns',66)
display(doc3)    #si noti come Pixel Draw - Number Art Coloring Book sia T in Art & Design e in Creativity
```

# 10) For each genre, compute the average rating. What is the genre with highest average?

```python
%%time

#utilizzo lo stesso procedimento dell'esercizio 8, con Rating al posto di App
rating = pd.DataFrame(doc['Rating'])

bg2 = pd.merge(rating, gen,left_index = True,right_index=True).melt(id_vars = ['Rating'], value_name = "Genres")
bg2=bg2.drop(columns = "variable").dropna().reset_index(drop=True)

display(bg2)
    
display(bg2.groupby('Genres').Rating.mean().sort_values(ascending=False))
#Le App di genere Events sono quelle con un rating medio maggiore.
```

# 11) For each app, compute the approximate income, obtain as a product of number of installs and price.

```python
%%time

doc.dtypes
m = re.compile('^(?P<dol>[$])*(?P<num>(\d+[.]*)*)$')

#definisco la funzione che converte il prezzo da stringa a valore numerico
def price2num(x):
    m1 = m.match(x)
    if m1 is None:
        return np.nan
    num = float(m1.group('num'))
    return num

doc.Price = doc.Price.apply(price2num)

doc.insert(loc = 8, column = 'Income', value = doc.Price * doc.Installs)   #per inserire income in una posizione specifica
display(doc.sort_values('Income',ascending=False))
```
